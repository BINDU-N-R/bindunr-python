#1.print multiples of 3 upto 200 using a list

```
L=[i for i in range(1,200) if i%3 ==0]
print(L)
```

#2. print the nos only divisible by 5 and 7 between 1000 and 2000 using a list(both inclusive)

```
L=[i for i in range(1000,2001) if i%5==0 and i%7==0]
print(L)
```

#3. Add the n number of names in a list and print them alphabetically and reverse alphabetically.

```
A=[]
n=int(input("Enter the n value : "))
for i in range(n):
  b=input("Enter {0} element : ".format(i+1))
  A.append(b)
A.sort()
print("Entered order is : ")
print(A)
A.reverse()
print("Reversed order is : ")
print(A)
```

#4. Print perfect squares and divisible by 5 between 500 and 1000 (both inclusive) using list comprehension

```
L=[i for i in range(500,1001) if (i**(.5) == int(i**(.5))and i%5==0)]
print(L)
```

#5. Calculate the Average of the numbers in a List by getting the elements of list from keyboard .

```
a=[]
n=int(input("Enter the number of elements in the list : "))
for i in range(0,n):
  L=int(input("Enter the {0} element : ".format(i+1)))
  a.append(L)
print(sum(a)/n)
```

#6. Print lists of odd,even and multiples of 5 numbers from 1 to 1000 using list comprehension

```
a=[i for i in range(5,1000,5)]
b=[i for i in range(1,1000)if (i%2)==0]
c=[i for i in range(1,1000)if (i%2)!=0]
print(a)
print(b)
print(c)
```

#7. Program to create a list of tuples with the first element as the numbers from 1 to 100 and the second element as the square root of the respective number.

```
import math
a=[(x,math.sqrt(x)) for x in range(1,100)]
print(a)
```
